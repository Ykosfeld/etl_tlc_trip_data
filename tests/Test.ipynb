```python
cabs = yellow_cab.unionByName(green_cab, allowMissingColumns=True)
```

```python
cabs.sample(0.0001).toPandas()
```

```python
from pyspark.sql.functions import *
```

```python
cabs = cabs.withColumn("pickup_datetime", coalesce("tpep_pickup_datetime", "lpep_pickup_datetime")).drop("tpep_pickup_datetime", "lpep_pickup_datetime")
cabs = cabs.withColumn("dropoff_datetime", coalesce("tpep_dropoff_datetime", "lpep_dropoff_datetime")).drop("tpep_dropoff_datetime", "lpep_dropoff_datetime")
```

```python
cabs.sample(0.0001).toPandas()
```

```python
cabs.filter(col("total_amount").isNull()).count()
```

```python
cabs.first()
```

```python
cabs = cabs.withColumn(
    "payment_type", 
    when(col("payment_type") == 5, None).otherwise(col("payment_type")))
```

```python
cabs.filter(col("payment_type")==5).count()
```

```python
cabs = cabs.withColumn(
    "RatecodeID",
    when(col("RatecodeID") == 99, None).otherwise(col("RatecodeID")))
```

```python
cabs.filter(col("RatecodeID").isNull()).count()
```

```python
cabs = cabs.withColumn(
        "Airport_fee",
        when(col("Airport_fee").isNull(), 0.0)
        .otherwise(col("Airport_fee"))
    )
```

```python
cabs.filter(col("Airport_fee")==0).count()
```

```python
cabs = cabs.withColumn(
        "congestion_surcharge",
        when(col("congestion_surcharge").isNull(), 0.0)
        .otherwise(col("congestion_surcharge"))
    )
```

```python
cabs.filter(col("congestion_surcharge")==0).count()
```

```python

for c in cabs.columns:
    n_nulls = cabs.filter(col(c).isNull()).count()
    print(f"{c}: {n_nulls}")
```

```python
cabs_deduped = cabs.dropDuplicates()
duplicate_count = cabs.count() - cabs_deduped.count()
```

```python
duplicate_count
```

```python
cabs = cabs.withColumn(
    "weekday_num",
    dayofweek(col("pickup_datetime"))
)
```

```python
cabs.first()
```

```python
cabs = cabs.withColumn(
    "trip_duration",
    expr("timestampdiff(SECOND, pickup_datetime, dropoff_datetime)")
)
```

```python
cabs.first()
```

```python
cabs = cabs.withColumn(
    "total_taxes",
    col("total_amount") - col("fare_amount")
)
```

```python
cabs.first()
```

```python
h = hour(col("pickup_datetime"))

cabs = cabs.withColumn(
    "hour_bucket",
    when(h.between(0, 6), 0)
        .when(h.between(7, 12), 1)
        .when(h.between(13, 18), 2)
        .when(h.between(19, 23), 3)
)
    
```
